# Print 

A função `print` é uma função imbutida do **Python**, que exibe mensagens na tela.

Quando você chama essa função, ele pega os argumentos fornecidos e exibe no console.

### Para que serve?

Serve para exibir mensagens, valores de variáveis, resultados de cálculos e outros tipos de dados no console (terminal)

```python
print('Olá, mundo!') # String entre aspas simples

print ("Olá, mundo!") # String entre aspas duplas

print (f'Olá, mundo!') # F-string é uma forma de formatar strings em Python 3.6+b
```

# Variáveis

---

### Definição

É um mecanismo de armazenamento volátil de dados, ou seja, dados salvos em pedacinhos da memória RAM do sistema operacional em uso (Notebook, Smartphone, Videogame, SmartTV, SmartWatch, entre tantas outras outras)

A criação de uma variável é realizada por meio da declaração e da atribuição de um valor a ela.

```python
nome = 'Eu sou Groot'

print ('Olá,', nome,', seja bem-vindo!') # Concatenação de Variável com sinal de ,
```

```python
print (f'Olá, {nome}, seja bem-vindo!') #F-string

print ('Olá, ' + nome + ', seja bem-vindo!') # Concatenação de Variável com sinal de +

print ('Olá', nome, 'seja bem-vindo', sep = '-')
```

```python
idade = 30

print(idade)

idade = 27

print(idade)
```

```python
print(nome.upper()) # Maiúscula

print(nome.lower()) # Minúscula

print(nome.capitalize()) # Somente o primeiro caractere em maiúscula, o restante em minúscula

print(nome.title()) # Somente a primeira letra de cada palavra em maiúscula

print(nome.swapcase()) # Inverte a capitalização dos caracteres {minuscula vira minuscula e vice-versa}
```
