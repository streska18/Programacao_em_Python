#### 5: Gritando o Código (Maiúsculas)

##### 5.1 Crie uma variável com o nome: codigo_secreto com o valor "missão alpha concluída".

##### 5.2 Use a função print() e o método apropriado para exibir o conteúdo da variável codigo_secreto totalmente em letras MAIÚSCULAS.

#### 7: Formatando um Início de Frase (Capitalize)

#### 8: Formatando um Título de Notícia (Title)

##### 8.1 Crie uma variável com o nome: noticia com o valor "próximo papa: muitos nomes, nenhum candidato oficial"

##### 8.2 Use a função print() e o método .title() para exibir o conteúdo da variável noticia formatado como um título (cada palavra começando com maiúscula).

```python
noticia = "próximo papa: muitos nomes, nenhum candidato oficial"

print(noticia.title())
```

#### 9: Invertendo a Caixa (Swapcase)

##### 9.1 Crie uma variável com o nome: texto_invertido com o valor "pROgrAmÇão PyThOn no sEnaI".

##### 9.2 Use a função print() e o método .swapcase() para exibir o conteúdo da variável texto_invertido com as maiúsculas e minúsculas trocadas.

```python
texto_invertido = "pROgrAmaÇão PyThOn no sEnaI"

print(texto_invertido.swapcase())
```
