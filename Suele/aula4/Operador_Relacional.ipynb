# Variáveis do Tipo Lógico

Muitas vezes queremos armazenar um conteúdo simples : Verdadeiro ou Falso em uma variável.
Neste caso, iremos utilizar a variável do tipo lógico (booleano)

Em **Python**, escrevemos *True* para Verdadeiro e *False* para Falso

Obs: O T e o F de True e False são em MAIÚSCULAS.

```python
resultado = True
aprovado = False

print(resultado)
print(f'Tipo da variável Resultado {type(resultado)}')
print(aprovado)
print(f'Tipo da variável Aprovado {type(aprovado)}')
```

# Operadores Relacionais

Para realizar comparações lógicas, utilizaremos operadores relacionais.
Os operadores de comparação lógica são:

- `>` Maior
- `>=` Maior ou Igual
- `<` Menor
- `<=` Menor ou Igual
- `==` Igual
- `!=` Diferente

```python
A = 1
B = 5
C = 2
D = 1
```

```python
A == B
```

```python
A < B
```

```python
A == D
```

```python
B >= A
```

```python
D != A
```

```python
# Exemplo de Caixa Eletrônico

saldo_em_conta = 200

valor_do_saque = 100

pode_executar_saque = valor_do_saque <= saldo_em_conta

print(pode_executar_saque)
```

```python
# Cartão de Crédito

cod_de_seguranca = '852'

cod_de_seguranca_cadastrado = '010'

pode_efetuar_pagamento = cod_de_seguranca == cod_de_seguranca_cadastrado

print(pode_efetuar_pagamento)
```
