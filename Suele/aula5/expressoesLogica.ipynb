# Expressões Lógicas

Os operadores lógicos, podem ser combinados em expressões lógicas mais complexas. Quando uma expressão tiver mais de um operador lógico, avalia-se o operador **not** primeiro, seguido do operador **and** e finalmente, operador **or**.

```python
# Operações booleanas

# Tabela Verdade do Operador Or (OR)

print (True or True)

print (True or False)

print (False or False)

print (False or True)
```

```python
# Tabela Verdade do Operador E (AND)

print (True and True)

print (True and False)

print (False and False)

print (False and True)

print (True and True and False)
```

```python
print (True or False and not True) # Inicia pelo not, depois executa as demais operações
```

```python
salario = int(input('Digite o Salário: '))
idade = int(input('Digite a Idade: '))

salario > 1000 and idade > 18
```

```python
# Tabela Verdade do Operador Não (NOT)

print (not(True and True))

print (not(True and False))

print (not(False and False))

print (not(False and True))

print (not(True and True and False))
```
