# Atividades - Estrutura Condicionais

```python
qtde_maca = int(input('Digite a quantidade de maçãs vendidas: '))
qtde_banana = int(input('Digite a quantidade de bananas vendidas: '))

if qtde_maca > qtde_banana:
    print ('As maçãs tiveram mais vendas!')
elif qtde_maca == qtde_banana:
    print ('Houve empate nas vendas!')
else:
    print ('As bananas tiveram mais vendas!')
```

```python
tempo_a = int(input('Digite o tempo para concluir a atividade A: '))
tempo_b = int(input('Digite o tempo para concluir a atividade B: '))
tempo_c = int(input('Digite o tempo para concluir a atividade C: '))

if tempo_a > 0 and tempo_b > 0 and tempo_c > 0:
    print (f'Total de dias para executar as atividades: {tempo_a + tempo_b + tempo_c}')
else:
    print ('Os dias não podem ser negativos. Não foi possível calcular!')
```

```python
temp = int(input('Digite a temperatura atual: '))

if temp > 25:
    print('Alerta! Temperatura acima do limite permitido.')
else:
    print('Temperatura dentro do permitido.')
```

```python
hora_entrada = int(input('Digite a hora de entrada (formato 24h): '))

if hora_entrada > 8 and hora_entrada < 18:
    print('Entrada permitida')
else:
    print('Entrada negada')
```

```python
distancia = int(input('Digite a distancia a ser percorrida: '))

if distancia < 100:
    print ('Valor do pedágio: R$10,00')
elif distancia >= 100 and distancia < 200:
    print ('Valor do pedágio: R$20,00')
else:
    print ('Valor do pedágio: R$30,00')
```

```python
renda_mensal = int(input('Digite a renda mensal: '))

parcela_desejada = int(input('Digite o valor da parcela que deseja pagar: '))

percentual = renda_mensal * 0.3

if renda_mensal > 2000 and parcela_desejada <= percentual:
    print ('Empréstimo liberado!')
else:
    print ('Empréstimo negado! Parcela acima de 30% da renda ou renda menor de R$2000,00')
    
```
