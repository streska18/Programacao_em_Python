# Abrindo o Sonho : Meu Negócio de Comida

```python
print('Passo 1')
print(' ')

ideia = input('Você possui uma ideia de comida original? Sim ou Não')
localizacao = input('Possui um ponto de venda confirmado? Sim ou Não')

if ideia.lower() == "sim" and localizacao.lower() == "sim":
    print ('Pronto para ir para o passo 2')
    print (" ")

    experiencia = float(input('Você sabe cozinhar minimamente? Se dê uma note de 0 a 10'))
    pesquisa_mercado = float(input('Você sabe o que as pessoas querem? Se dê uma nota de 0 a 10'))
    devedor = int(input('Você possui quantos devedores?'))

    if experiencia > 5 and pesquisa_mercado > 0 and devedor < 3:
        print ('Pronto para ir para o próximo passo 3')
        print (" ")

        economia_propria = input('Você possui economia própria? Responda sim ou não')
        emprestimo = input ('Você possui algum amigo ou familia que possa te emprestar dinheiro? Responda sim ou não')
        credito = input ('Você possui crédito de pequenos negócios de um banco ou uma cooperativa de crédito? Responda sim ou não')
        financiamento = input('Você possui alguma ajuda com financiamento coletivo? Ex.: Vaquinha online. Responda sim ou não')

        if economia_propria.lower() == "sim" or emprestimo.lower() == "sim" or credito.lower() == "sim" or financiamento.lower() == "sim":
            print ("Parabéns, você está pronto para abrir o seu negócio!")
        else:
            print ('Não é possível iniciar seu negócio!')

    else:
         print ('Não funciona!')

else:
    print ('Não funciona!')
```
